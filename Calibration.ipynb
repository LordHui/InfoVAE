```python
os.environ['CUDA_VISIBLE_DEVICES'] = '3'
dataset = MnistDataset(binary=True)
c = GenerativeAdversarialNet(dataset, name='wgan_mnist')
c.train(100)
```

```python
def calibrate(x_sample):
    name = 'calibration'
    c_lb = tf.get_variable('%s_lb' % name, shape=[28, 28, 1], dtype=tf.float32, initializer=tf.constant_initializer(0.3))
    c_ub = tf.get_variable('%s_ub' % name, shape=[28, 28, 1], dtype=tf.float32, initializer=tf.constant_initializer(0.7))
    x_mean = x_sample * tf.minimum(c_ub, 0.995) + (1 - x_sample) * tf.maximum(c_lb, 0.005)
    return x_mean, [c_lb, c_ub]

def calibrate_conv(x_sample):
    with tf.variable_scope('calib') as vs:
        conv1 = conv2d_lrelu(x_sample, 16, 3, 1)
        x_mean = tf.contrib.layers.conv2d(conv1, 1, 3, 1, activation_fn=tf.sigmoid)
        x_mean = tf.maximum(tf.minimum(x_mean, 0.995), 0.005)
    return x_mean, [var for var in tf.global_variables() if 'calib' in var.name]
```

```python
def variational_posterior(x, z_dim):
    with tf.variable_scope('vi_posterior'):
        conv1 = conv2d_lrelu(x, 64, 4, 2)   # None x 14 x 14 x 64
        conv2 = conv2d_lrelu(conv1, 128, 4, 2)   # None x 7 x 7 x 128
        conv2 = tf.reshape(conv2, [-1, np.prod(conv2.get_shape().as_list()[1:])]) # None x (7x7x128)
        fc1 = fc_lrelu(conv2, 1024)
        mean = tf.contrib.layers.fully_connected(fc1, z_dim, activation_fn=tf.identity)
        stddev = tf.contrib.layers.fully_connected(fc1, z_dim, activation_fn=tf.sigmoid)
        stddev = tf.maximum(stddev, 0.01)
        return mean, stddev
```

```python
x = tf.placeholder(shape=[None, 28, 28, 1], dtype=tf.float32)
z_mean, z_stddev = variational_posterior(x, 10)
z_sample = z_mean + tf.multiply(z_stddev,
                                tf.random_normal(tf.stack([tf.shape(x)[0], 10])))
```

```python
x_sample = c.generator(z_sample, data_dims=c.data_dims, reuse=True)
x_mean, calib_var = calibrate_conv(x_sample)
```

```python
nll_per_sample = -tf.reduce_sum(tf.log(x_mean) * x + tf.log(1 - x_mean) * (1 - x), axis=(1, 2, 3))
loss_nll = tf.reduce_mean(nll_per_sample)

# ELBO loss divided by input dimensions
loss_elbo_per_sample = tf.reduce_sum(-tf.log(z_stddev) + 0.5 * tf.square(z_stddev) +
                                     0.5 * tf.square(z_mean) - 0.5, axis=1)
loss_elbo = tf.reduce_mean(loss_elbo_per_sample)
```

```python
is_estimator = nll_per_sample + loss_elbo_per_sample
```

```python
vi_var = [var for var in tf.global_variables() if 'vi_posterior' in var.name]
with tf.variable_scope('adam_train'):
    calib_train = tf.train.AdamOptimizer(1e-4).minimize(loss_nll + loss_elbo, var_list=calib_var + vi_var)
    vi_train = tf.train.AdamOptimizer(1e-4).minimize(loss_nll + loss_elbo, var_list=vi_var)
adam_var = [var for var in tf.global_variables() if 'adam_train' in var.name]
```

```python
c.sess.run(tf.variables_initializer(vi_var + calib_var + adam_var))
for iter in range(10000):
    bx = c.dataset.next_batch(100)
    nll, elbo, _ = c.sess.run([loss_nll, loss_elbo, calib_train], feed_dict={x: bx})
    if iter % 100 == 0:
        test_bx = c.dataset.next_test_batch(100)
        test_nll, test_elbo = c.sess.run([loss_nll, loss_elbo], feed_dict={x: test_bx})
        print("Iteration %d, nll=%.4f/%.4f, elbo=%.4f/%.4f" % (iter, nll, test_nll, elbo, test_elbo))
    if iter % 1000 == 0:
        mean = c.sess.run(x_mean, feed_dict={x: bx})
        for i in range(5):
            plt.subplot(5, 2, 2*i+1)
            plt.imshow(bx[i, :, :, 0], vmin=0, vmax=1, cmap='Greys')
            plt.subplot(5, 2, 2*i+2)
            plt.imshow(mean[i, :, :, 0], vmin=0, vmax=1, cmap='Greys')
            plt.show()
```

```python
c.sess.run(tf.variables_initializer(vi_var + [var for var in tf.global_variables() if 'vi_train' in var.name]))
for iter in range(10000):
    bx = c.dataset.next_test_batch(100)
    nll, elbo, _ = c.sess.run([loss_nll, loss_elbo, vi_train], feed_dict={x: bx})
    if iter % 100 == 0:
        print("Iteration %d, nll=%.4f, elbo=%.4f" % (iter, nll, elbo))
```

```python
# Evaluate the log likelihood on test data
def compute_log_sum(val):
    min_val = np.min(val, axis=0, keepdims=True)
    return np.mean(min_val - np.log(np.mean(np.exp(-val + min_val), axis=0)))


def compute_nll_by_is(batch_x, verbose=False):
    start_time = time.time()
    nll_list = []
    num_iter = 2000
    for k in range(num_iter):
        nll = c.sess.run(is_estimator, feed_dict={x: batch_x})
        nll_list.append(nll)
        if verbose and k % 500 == 0:
            print("Iter %d, current value %.4f, time used %.2f" % (
                k, compute_log_sum(np.stack(nll_list)), time.time() - start_time))
    return compute_log_sum(np.stack(nll_list))
```

```python
train_nll = []
test_nll = []
for i in range(50):
    batch_x = c.dataset.next_test_batch(100)
    nll = compute_nll_by_is(batch_x)
    train_nll.append(nll)
    batch_x = c.dataset.next_batch(100)
    nll = compute_nll_by_is(batch_x)
    test_nll.append(nll)
    print("Nll is %.4f/%.4f" % (np.mean(train_nll), np.mean(test_nll)))
```
