```python
from matplotlib import pyplot as plt
```

```python
names = ['log_mmd', 'log_0.000000', 'log_0.010000', 'log_0.001000']
c_list = ['r', 'g', 'b', 'm']
for name, c in zip(names, c_list):
    step, val = read_file(name)
    plt.plot(step, val, c=c, hold=True)
plt.xscale('log')
plt.ylim([0, 0.07])
plt.show()
```

```python
def read_file(filename):
    reader = open(filename, 'r')
    step_list, val_list = [], []
    while True:
        line = reader.readline().split()
        if len(line) < 2:
            break
        step_list.append(int(line[0]))
        val_list.append(float(line[1]))
    return step_list, val_list
```
